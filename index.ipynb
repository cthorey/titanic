# Diversity of the titanic population

Instead of taking a morbid approach to the problem right away, I am first going to look at the diversity of the titanic population before the tragedy (assuming the training set contain a representative sample of the whole population). To do this, I am first going to clean up the data a bit based on a few assumptions:
- LastName can be used as a proxy for Sibling/Parents..
- PassengerId does not bring any useful info
- I am going to assume that Fare and Pclass are redundant and only look at Pclass
- Assume that ticket does not bring any useful info.
- Cabin is mostly Nan and I am going to remove it as well.

```python
data = data[[f for f in data.columns if f not in ['Cabin','PassengerId']]].dropna()
# extract LastName
data['LastName'] = data.Name.map(lambda x:x.split(',')[0])
# convert Age as int
data['Age'] = [f for f in map(lambda x:int(x),data.Age)]
df = data[[f for f in data.columns if f in ['Survived','Pclass','Sex','Age','Embarked','LastName']]]
df.info()
```

Let's also encoded Sex,Embarked and LastName as integer to ease further processing using the labelEncoder class of sklearn

```python
# lsex - female:1,male:0
lsex = LabelEncoder().fit(df.Sex)
# lembarked - Cherbourg:0, Queenstown:1, Southampton:2 
lembarked = LabelEncoder().fit(df.Embarked)
llastname = LabelEncoder().fit(df.LastName)

df.loc[:,'Sex'] = lsex.transform(df.Sex)
df.loc[:,'Embarked'] = lembarked.transform(df.Embarked)
df.loc[:,'LastName'] = llastname.transform(df.LastName)
```

## Age distribution

Let's begin by drawing the normalized age distribution of the passenger (in the train set..)
- It is bimodal with one mode around baby/kids and another one around 30 years old. 
- It also display a thick slowly decreasing tail up to 80 years old

It is expected assuming we have 
- 1) old rich people contributing to the heavy tail
- 2) families contributing to the first mode and second mode
- 3) lonely travellers (mostly male ?) contributing to the second mode.

```python
fig = plt.figure(figsize=(12,6))
ax = plt.subplot()
sns.distplot(data.Age,rug=True,kde=False,ax=ax,norm_hist=True)
ax.tick_params(labelsize=22)
ax.set_xlabel('Age',fontsize=22)
ax.set_ylabel('Frequency',fontsize=22)
```

To confirm these hypotheses, let's look at the distribution of age by passenger class. This is a reasonably good proxy for socio economic factor. 

```python
import seaborn as sns
fig = plt.figure(figsize=(12,6))
ax = plt.subplot(111)
# Draw a swarm plot summarizing the distribution for each passenger class independently.
sns.swarmplot(x="Pclass", y="Age", data=data,ax=ax,
               orient='v',split=True)
sns.despine(left=True)
ax.set_ylim(0,100)
ax.tick_params(labelsize=22)
ax.set_ylabel('Age',fontsize=22)
ax.set_xlabel('Pclass',fontsize=22)
ax.set_title('Age distribution by passenger class',fontsize=22)
```

## Lonely travellers

It's quite a lot of them.

## Rapid modelling 

To finish this rapid exploration, let's try to do a simple logistic regression based on the Age, class, city and nb of family number on board. Given this regressor, we might already be able to get a good score on the leaderboard with a simple linear model !

```python
import statsmodels.api as sm
```

```python
model = sm.Logit(lonely['Survived'], lonely[['Age','Pclass','Embarked','CntLastname']])
result = model.fit()
```

```python
print(result.summary())
```

It appears that a simple logistic regression does not have a lot of predictive power for this dataset. The pseudo R-squared is very small and the hability of this model to explain the survival rate is not much better than a model without any regressors... 

Still, it suggests that

- The odds of surviving decrease slightly with Age. Same for going from 1st class to 3rd class and for embarking from South. instead of Cherbourg.
- But, if you ever wanted to survived in a boat disaster sometime in the future, takes a lot of sibling/parents with you.

The coefficient for math says that, holding ther variable constant, we will see a 93% decrease in the odds of surviving for a one-unit increase in Age.

```python
# odds ratios and 95% CI
params = result.params
conf = result.conf_int()
conf['OR'] = params
conf.columns = ['2.5%', '97.5%', 'OR']
print(np.exp(conf))
```
