# Scratching the nature of the titanic population

Instead of taking a morbid approach to the problem right away, I am first going to understand the diversity of the titanic population before the tragedy (assuming the training set contain a representative sample of the whole population). To do this, I am first to clean up the data a bit based on a few assumptions:
- LastName can be used as a proxy for Sibling/Parents..
- PassengerId does not bring any useful info
- I am going to assume that Fare and Pclass are redundant and only look at Pclass
- Assume that ticket does not bring any useful info.
- Cabin is mostly Nan and I am going to remove it as well.

```python
data = data[[f for f in data.columns if f not in ['Cabin','PassengerId']]].dropna()
# extract LastName
data['LastName'] = data.Name.map(lambda x:x.split(',')[0])
# convert Age as int
data['Age'] = [f for f in map(lambda x:int(x),data.Age)]
df = data[[f for f in data.columns if f in ['Survived','Pclass','Sex','Age','Embarked','LastName']]]
df.info()
```

Let's also encoded Sex,Embarked and LastName as integer to ease further processing using the labelEncoder class of sklearn

```python
# lsex - female:1,male:0
lsex = LabelEncoder().fit(df.Sex)
# lembarked - Cherbourg:0, Queenstown:1, Southampton:2 
lembarked = LabelEncoder().fit(df.Embarked)
llastname = LabelEncoder().fit(df.LastName)

df.loc[:,'Sex'] = lsex.transform(df.Sex)
df.loc[:,'Embarked'] = lembarked.transform(df.Embarked)
df.loc[:,'LastName'] = llastname.transform(df.LastName)
```

## Age distribution

Let's begin by drawing the normalized age distribution of the passenger (in the train set..)
- It is bimodal with one mode around baby/kids and another one around 30 years old. 
- It also display a thick slowly decreasing tail up to 80 years old

It is expected assuming we have 
- 1) old rich people contributing to the heavy tail
- 2) families contributing to the first mode and second mode
- 3) lonely travellers (mostly male ?) contributing to the second mode.

```python
fig = plt.figure(figsize=(12,6))
ax = plt.subplot()
sns.distplot(data.Age,rug=True,kde=False,ax=ax,norm_hist=True)
ax.tick_params(labelsize=22)
ax.set_xlabel('Age',fontsize=22)
ax.set_ylabel('Frequency',fontsize=22)
```

To confirm these hypotheses, let's look at the distribution of age by passenger class. This is a reasonably good proxy for socio economic factor. 

```python
import seaborn as sns
fig = plt.figure(figsize=(12,6))
ax = plt.subplot(111)
# Draw a swarm plot summarizing the distribution for each passenger class independently.
sns.swarmplot(x="Pclass", y="Age", data=data,ax=ax,
               orient='v',split=True)
sns.despine(left=True)
ax.set_ylim(0,100)
ax.tick_params(labelsize=22)
ax.set_ylabel('Age',fontsize=22)
ax.set_xlabel('Pclass',fontsize=22)
ax.set_title('Age distribution by passenger class',fontsize=22)
```

```python
df.groupby('Pclass').Age.mean()
```

At this point we cannot really draw any conclusion on family Vs lonely travellers. Let's take a deeper look to the lonely traveller population. We can isolate using a slightly oversimplifying  assumption, here it is:
- Group people by their last name and assume that people on board that are uniquely identified by this one are alone !

## Lonely travellers

```python
cnt = df.groupby('LastName').size().to_dict()
df['CntLastname'] = [f for f in map(lambda x:cnt[x], df.LastName)]
#ax = (data.CntLastname).hist(figsize=(16,8))
sns.distplot(df.CntLastname,kde=False)
ax.tick_params(labelsize=22)
ax.set_xlim(1,9)
ax.set_xlabel('Nb of persons appearing with a specific last name',fontsize=22)
ax.set_ylabel('Count',fontsize=22)
```

It's quite a lot of them.

```python
lonely = df[df.CntLastname==1]
fig = plt.figure(figsize=(12,6))
ax = plt.subplot()
sns.distplot(lonely.Age,rug=True,kde=False,ax=ax,norm_hist=True)
ax.tick_params(labelsize=22)
ax.set_xlabel('Age',fontsize=22)
ax.set_ylabel('Frequency',fontsize=22)
```

It looks like lonely babies were abandonned on the titanic ... I said it was an oversimplifying assumption. That said, the age distribution looks now much more unimodal. From there, we can ask several things:

As expected.
- The likelihood of survival is large (~1) for, again "lonely", kids younger than 15 and passenger older than 50. 
- Young fellow had a very low likelihood of survival. And it is biased toward large values due to the relatively high survival rate the 1st class passenger. If we remove them from the pool, it got worst...

```python
fig = plt.figure(figsize=(23,8))
ax = plt.subplot()
sns.barplot(x="Age", y="Survived", data=lonely[lonely.Pclass.isin([2,3])],
            ax=ax,ci=95,color=sns.color_palette('deep')[0])
ax.set_ylabel('Rate of survival',fontsize=22)
ax.set_xlabel('Age',fontsize=22)
ax.tick_params(labelsize=22)
plt.xticks(rotation=90)
```

## Modelling 

```python
import pymc3 as pm
```

```python
rate = (lonely.groupby('Age').Survived.sum()/
        lonely.groupby('Age').Survived.size()).reset_index(name='rate')
```

```python
with pm.Model() as model:
    theta_prior = pm.Uniform('theta',lower=0,upper=1)
    likelihood = pm.Bernoulli('y_est',theta_prior,observed=rate.rate.tolist())
    
    start = pm.find_MAP()
    step = pm.Metropolis()
    trace = sample(2000, step, start=start, progressbar=True)
```
